```python
# Import libraries
import pandas as pd

# Load the dataset
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("/content/Superstore_Cleaned.csv")

# Dropping the column with missing values
df.dropna(inplace=True)


# Saved the cleaned data
df.to_csv("cleaned_superstore.csv", index=False)

# Calculating the key metrics
top_categories = df.groupby('Category')['Profit'].sum().nlargest(3)
total_sales = df['Sales'].sum()

# Print results
print("🚀 Automated Sales Analysis Results:")
print(f"Total Sales: ${total_sales:,.2f}")
print("\nTop 3 Profitable Categories:")
print(top_categories)
```
