```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Superstore_Cleaned.csv')
```

```python
## top 3 categories from superstore by profit

top_categories = df.groupby('Category')['Profit'].sum().nlargest(3).reset_index()
print("Top 3 Profitable Categories:\n", top_categories)
```

```python
## sales performance by region.

region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False).reset_index()
print("Sales by Region:\n", region_sales)
```

```python
print(df.columns)
```

```python
## top 5 customers by amount spending


top_customers = df.groupby('Customer_ID')['Sales'].sum().nlargest(5).reset_index()
print("Top 5 Customers by Spend:\n", top_customers)
```

 Profit by Category (Bar Plot)

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Profit', data=df, estimator=sum, ci=None)
plt.title("Total Profit by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=df, hue='Category')
plt.title("Sales vs. Profit (by Category)")
plt.show()
```
